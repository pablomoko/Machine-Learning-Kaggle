```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')

train_data = pd.read_parquet("/content/drive/MyDrive/TP2 - ORGANIZACIÃ“N DE DATOS/dataset_v2/train.parq", engine="pyarrow")
train_data.dropna(inplace=True)
```

## Frecuencia de Ataques por Tipo y Target

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(x='attack_type', hue='label', data=train_data)
plt.xticks(rotation=45)
plt.title('Frecuencia de Ataques por Tipo y Target')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar los datos donde label es igual a 1
df_label_1 = train_data[train_data['label'] == 1]

plt.figure(figsize=(10, 6))
sns.countplot(x='attack_type', data=df_label_1)
plt.xticks(rotation=45)
plt.title('Frecuencia de Ataques por Tipo con label igual a 1')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar los datos donde label es igual a 1
df_label_1 = train_data[train_data['label'] == 0]

plt.figure(figsize=(10, 6))
sns.countplot(x='attack_type', data=df_label_1)
plt.xticks(rotation=45)
plt.title('Frecuencia de Ataques por Tipo con label igual a 0')
plt.show()
```
