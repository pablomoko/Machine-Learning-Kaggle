## Distribución de Tipos de Ataque por Día de la Semana (solo label 1)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Filtrar los datos donde label es igual a 1
df_label_1 = train_data[train_data['label'] == 1]

# Obtener el día de la semana
day_of_week = df_label_1['attack_time'].dt.day_name()

# Crear el heatmap
plt.figure(figsize=(12, 8))
heatmap_data = pd.crosstab(day_of_week, df_label_1['attack_type'], normalize='index')
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt=".2%", cbar_kws={'label': 'Porcentaje'})
plt.title("Distribución de Tipos de Ataque por Día de la Semana (solo label 1)")
plt.xlabel("Tipo de Ataque")
plt.ylabel("Día de la Semana")
plt.show()
```
