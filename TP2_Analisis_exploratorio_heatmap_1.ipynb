## Relación entre tipo de ataque y label (normalizados)

Dado al alto desequilibrio, quiero mostrar si hay alguna relacion entre los ataques y el label

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Normalizar los datos por etiqueta
normalized_data = train_data.groupby(['label', 'attack_type']).size().unstack()
normalized_data = normalized_data.div(normalized_data.sum(axis=1), axis=0)  # Normalización por filas

plt.figure(figsize=(10, 6))
sns.heatmap(normalized_data, cmap='YlOrRd', annot=True, fmt='.2f', linewidths=1, linecolor='black')
plt.title("Relación entre tipo de ataque y label (normalizados)")
plt.xlabel("Tipo de Ataque")
plt.ylabel("Label")
plt.show()
```
